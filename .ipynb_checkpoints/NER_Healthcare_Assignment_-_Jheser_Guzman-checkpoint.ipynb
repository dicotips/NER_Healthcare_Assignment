### Auxiliar Funcion to Unpack files

To pre-process the given text corpus and transform it to sentences from list of words (inline), I have created the following function that takes the file name as input and unpacks line by line to form sentences.
It returns the List of sentences to the calling module.

```python
def unpack_file(file_name):
    ''' Function that takes the filename as input and unpacks line by line to form sentences.
        It returns the List of sentences'''
    
    file = open(file_name,'r')
    data = file.readlines()
    file.close()
    L = [];
    sent = []
    for line in data:
        if line == '\n':
            sent.append(" ".join(L))
            L=[]
        else: 
            L.append(line.rstrip('\n')) # Keep adding to the temporary list
    return sent
```

## Defining features for CRF

Computing Features (f) using the following features for our input sequences.
We have defined the following features for CRF model building:

* f1 = postag of the word
* f2 = word in lower case
* f3 = last 3 characters of the word
* f4 = last 2 characters of the word
* f5 = 1; if the word is in uppercase, otherwise 0
* f6 = 1; if word is a number; otherwise 0
* f7 = 1; if the word starts with a capital letter, otherwise 0

* If the given word is not the beginning of the sentence, we include the features of the previous word too
* If the word is at the beginning of sentence we mark it as 'BEG'
* If the word is at the end of the sentence we mark it as 'END'

```python
# Let's define the features to get the feature value for a keyword.
def getFeaturesForAWord(sentence, pos):
  word = sentence[pos]
  for token in model(word):
    word_pos_tag = token.pos_
  features=[
    'word.postag=' + word_pos_tag,                   # f1: PoS tag of the word
    'word.lower=' + word.lower(),                    # f2: input word is in lower case
    'word[-3:]=' + word[-3:],                        # Last three characters
    'word[-2:]=' + word[-2:],                        # Last two characters
    'word.isupper=%s' % word.isupper(),              # Is word in all upper case
    'word.isdigit=%s' % word.isdigit(),              # Is word a number
    'words.startsWithCapital=%s' % word[0].isupper() # Is word starting with capital letter
  ]

  if pos > 0:
    prev_word = sentence[pos-1]
    for token in model(prev_word):
      prev_word_pos_tag = token.pos_
    features.extend([
      'prev_word.postag=' + prev_word_pos_tag,      # Previous word PoS tag
      'prev_word.lower=' + prev_word.lower(),       # Previous word in lower case
      'prev_word.isupper=%s' % prev_word.isupper(), # Is previous word in upper case
      'prev_word.isdigit=%s' % prev_word.isdigit(), # Is previous word a number
      'prev_words.startsWithCapital=%s' % prev_word[0].isupper() # is previous word starting with capital letter
    ])
  else:
    features.append('BEG') # Beginning of sentence
  if (pos == len(sentence) - 1):
    features.append('END') # End of sentence
  return features
```

## Define Input and Target Variables

Correctly computing X and Y sequence matrices for training and test data.
Check that both sentences and labels are processed

### Define the features' values for each sentence as input variable  for CRF model in test and the train dataset 

```python
X_train = [getFeaturesForASentence(sentence) for sentence in train_set]
Y_train = [getLabelsInListForASentence(labels) for labels in train_label]
```

### Define the labels as the target variable for test and the train dataset

```python
X_test = [getFeaturesForASentence(sentence) for sentence in test_set]
Y_test = [getLabelsInListForASentence(labels) for labels in test_label]
```

## Build the CRF Model

```python
crf = sklearn_crfsuite.CRF( max_iterations = 100)
```

```python
crf.fit(X_train, Y_train)
```

```python
## AttributeError: 'CRF' object has no attribute 'keep_tempfiles'
# This error is related to the Sckit-Learn version that does not keep tempfiles. The model training is not affected by the
# error, so we can continue with the trained model
```

## Evaluating the Model

### Predict the labels of each of the tokens in each sentence of the test dataset that has been pre processed earlier.

```python
Y_pred = crf.predict(X_test)
```

### Calculate the f1 score using the actual labels and the predicted labels of the test dataset.

```python
metrics.flat_f1_score(Y_test, Y_pred, average='weighted')
```

#### We have got the f1 score of 0.9054

## Identifying Diseases and Treatments using Custom NER

We now use the CRF model's prediction to prepare a record of diseases identified in the corpus and treatments used for the diseases.



### Create the logic to get all the predicted treatments (T) labels corresponding to each disease (D) label in the test dataset.

image.png

```python
# Creating a dictionary out of the predicted labels from our model data
dict = {}
for count, labels in enumerate(Y_pred):
    temp = "".join(labels)                 # removing the spaces between the lablels
    totalD = len(re.findall('D', temp))    # find number of 'D' in the sentence
    totalT = len(re.findall('T', temp))    # find number of 'T' in the sentence
    if totalD > 0 and totalT > 0:
        indexD=[];
        indexT=[];

        start_from = 0
        for i in range(totalD):
            foundat = temp.find('D', start_from)
            indexD.append(foundat)
            start_from = foundat+1

        start_from = 0
        for i in range(totalT):
            foundat = temp.find('T', start_from) # search for 'T' from start_from in the labels
            indexT.append(foundat)
            start_from = foundat + 1

            
        disease = ''
        for i in indexD: # Looping through the indexes where 'D' is located
            disease = disease + test_set[count].split()[i]
            if i+1 not in indexD:
                disease = disease + ', '
            else:
                disease = disease + ' '

        treatment = ''
        for i in indexT: # Looping through the indexes where 'D' is located
            treatment = treatment + test_set[count].split()[i]
            if i+1 not in indexT:
                treatment = treatment + ', '
            else:
                treatment = treatment + ' '
        
        # Adding the disease and treatment in the dictionary
        dict[disease.rstrip(', ')]=treatment.rstrip(', ') 
            
dict
```
